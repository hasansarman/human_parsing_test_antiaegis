```python
from matplotlib import pyplot as plt
from data_loaders_motion import BinarySegmentationDataLoader
import numpy as np

import sys
sys.path.append("../")
```

```python
data_loader = BinarySegmentationDataLoader(
    "../dataset/mask_averg_train_pairs.txt",
    color_channel="RGB",
    resize=224,
    padding_value=0,
    ignore_index=255,
    noise_std=5,
    n_augment=7,
    is_training=True,
    batch_size=1, 
    n_workers=1,
    pin_memory=True,
    shuffle=True,
).loader
```

```python
for images, labels in data_loader:
    plt.figure(1, figsize=(12,6*len(images)))
    for i in range(len(images)):
        image = images[i,...].cpu().numpy()
        image = np.transpose(image, [1,2,0])
        label = labels[i,...].cpu().numpy()
        print(label.shape)

        plt.subplot(len(images), 2, 2*i+1)
        plt.imshow(image)
        plt.subplot(len(images), 2, 2*i+2)
        plt.imshow(image)
        plt.imshow(label, alpha=0.5)

    plt.show()
    break
```
